```python
%matplotlib notebook
%config InlineBackend.figure_format = 'svg'
#import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# get column names
SS = pd.read_csv('net_survival_by_primary_site.csv')
SS.columns
SS
```

```python
NC= ['Leukemia',
 'Liver and intrahepatic bile duct',
 'Brain and other nervous system',
 'Pancreas','Esophagus',
 'Lung and bronchus',
 'Myeloma',
 'Stomach',
 'Urinary bladder',
 'Larynx',
 'Oral cavity and pharynx',
 'Non-Hodgkin lymphoma',
 'Ovary',
 'Cervix uteri',
 'Hodgkin lymphoma',
 'Colon and rectum',
 'Kidney and renal pelvis',
 'Corpus and uterus, NOS',
 'Melanoma of the skin',
 'Testis',
 'Breast',
 'Thyroid',
 'Prostate']
SS['cancer'] = NC


SS
```

```python
df = SS.loc[:,['cancer','stage_group','survival','lower_bound','upper_bound']]
#dff = df.set_index(['cancer','stage_group'])
#sns.barplot(y = 'cancer', x = 'survival', hue="stage_group", data = df, errwidth= 0.3 ,orient = 'h')
#sns.barplot(x="day", y="total_bill", data=tips)
df['cancer'] = df['cancer'].astype('category')
df['lower']= df['survival'] - df['lower_bound']
df['upper']= df['upper_bound'] - df['survival'] 
df
```

```python
print(plt.style.available)
plt.style.use('seaborn-white')

plt.figure(figsize=(8,8))
plt.errorbar(y ='cancer', x ='survival', xerr=(df['lower'], df['upper']), fmt = 'bo', alpha = 0.5 , data = df )


plt.subplots_adjust(wspace=0.2, left=0.27, bottom=0.1, right= 0.95, top=0.9)
#plt.subplots_adjust(wspace=0.2, hspace=0.7, left=0.3, bottom=0.22, right=0.8, top=1.2)
plt.gca().set_xticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_xticks()]) 


#plt.legend(loc = 8, frameon = False)
#plt.ylabel('cancer')
plt.xlabel('Net survival (%)')

plt.title('Five-year net survival of cancer patients in Pennsylvana \n aged 15+, by primary site \n')
ax = plt.axes()        
ax.xaxis.grid()

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.show()
```
